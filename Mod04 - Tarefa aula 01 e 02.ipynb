#### 1)  crie uma série do pandas a partir de uma lista com os dados abaixo:

Em um estudo sobre alteração na temperatura global, A NASA disponibiliza dados de diferenças de temperatura média da superfície terrestre relativo às médias de temperatura entre 1951 e 1980. Os dados originais podem ser vistos no site da NASA/GISS, e estão dispostos a cada década na tabela abaixo.

|ano|anomalia termica|
|:-:|:----:|
| 1900 | -0.08 |
| 1920 | -0.27 |
| 1940 | 0.12 |
| 1960 | -0.03 |
| 1980 | 0.26 |
| 2000 | 0.40 |
| 2020 | 1.02 |

Crie uma séries do Pandas a partir de uma lista com esses dados.

```python
import pandas as pd
import numpy as np

list_anos = ('1900', '1920', '1940', '1960', '1980', '2000', '2020')
list_anomalias = (-0.08, -0.27, 0.12, -0.03, 0.26, 0.4, 1.02)


serie_1 = pd.Series(list_anos, list_anomalias)
serie_1
```

#### 5) Nomeie os índices das linhas com inteiros de 1 a 20, e as colunas com os nomes "x1", "x2", e "x3" respectivamente.

```python
arr_enumerado = pd.DataFrame(arr, columns=['x1', 'x2', 'x3']) # Nomeando as colunas
arr_enumerado.index = range(1, 21) # nomeando as linhas com o método range
arr_enumerado
```

#### 6) No DataFrame do exercício 5, crie uma nova coluna como sendo a média das três colunas, e dê a ela o nome de "media" (não recomendo colocar acentos em nomes de variáveis).

```python
arr_enumerado['Media'] = arr_enumerado['x1'] * arr_enumerado['x2'] * arr_enumerado['x3'] / 3
arr_enumerado
```
